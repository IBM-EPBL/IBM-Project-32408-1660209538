```python
#Taking the open column for training
train_set = df.iloc[:,1:2].values
train_set
```

```python
# Scaling

from sklearn.preprocessing import LabelEncoder

sc = LabelEncoder()
scaled_tr = sc.fit_transform(train_set)
scaled_tr
```

```python
sc.inverse_transform([[1079]])
```

```python
scaled_tr[60-60:60-0]
```

CREATE THE MODEL

```python
# Split dependent and independent

x_train = []
y_train = []

for i in range(60,1258):
  x_train.append(scaled_tr[i-60:i-0])
  y_train.append(scaled_tr[i-0])
```

```python
x_train
```

```python
x_train,y_train = np.array(x_train),np.array(y_train)
```

```python
x_train
```

```python
x_train.shape
```

```python
x_train = np.reshape(x_train,(1198, 60, 1))
x_train.shape
```

TEST THE MODEL

```python
x_test = []
y_test = []

for i in range(60,1258):
  x_test.append(scaled_tr[i-60:i-0])
  y_test.append(scaled_tr[i-0])
```

```python
x_test
```

ADDING LAYERS

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM
```

```python
model = Sequential()
model.add(LSTM(50, input_shape = (60,1),return_sequences=True))
model.add(LSTM(50,return_sequences=True))
model.add(LSTM(50,return_sequences=True))
model.add(LSTM(50,return_sequences=True))
model.add(Dense(1))
```

COMPILE THE MODEL

```python
model.compile(optimizer='adam',loss='mse')
```

FIT THE MODEL

```python
model.fit(x_train, y_train, epochs=10, batch_size=32)
```

SAVE THE MODEL

```python
model.save("Assignment5.h5")
```
